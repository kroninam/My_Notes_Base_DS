# Повторяем способы достать данные из строк и колонок:

```python
df.reindex(['gender', 'race/ethnicity'], axis = 1)[:3]
```

```python
df.loc[[0,1,2], ['gender', 'race/ethnicity']]
```

```python
# множественные запросы оборачиваем в дополнительный список:

df.iloc[[0, 1, 10], [0, 5]]
```

```python
df.iloc[:3, :2]
```

```python
# practice:

# df.iloc[[0, 3, 4, 7, 8], :]
df.iloc[[0, 3, 4, 7, 8]]
```

```python
df_1 = df.iloc[[0, 3, 4, 7, 8]]
```

```python
df_1.index = ["Cersei", "Tywin", "Gregor", "Joffrey", "Ilyn Payne"]
```

```python
df_1
```

```python
df_1.loc[['Cersei'], ['gender']]
```

```python
df_1.loc[['Cersei', 'Joffrey']]
```

```python
# DataFrame состоит из Series:

s1 = pd.Series([1,2,3], index = ["Cersei", "Tywin", "Gregor"])
s2 = pd.Series([4,5,6], index = ["Cersei", "Tywin", "Gregor"])

df = pd.DataFrame({'col_1': s1, 'col_2': s2})

df
```

```python
df.iloc[:7]
```

```python
df.loc[:6]
```
