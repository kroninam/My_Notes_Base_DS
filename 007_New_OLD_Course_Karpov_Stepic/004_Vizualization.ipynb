```python
df['count_roles'] = df['roles'].apply(lambda x: len(x.split(',')))
```

```python
# самое часто встречающееся - это "мода", в датафрейме функция .mode() !!!

df['count_roles'].mode()
```

```python
int(df['count_roles'].mode().values)
```

```python
df['mode'] = int(df['count_roles'].mode().values)
```

```python
sns.barplot(x='id', y='count_roles', palette = 'dark', data = df)
sns.lineplot(x="id", y="mode", data=df)
```
