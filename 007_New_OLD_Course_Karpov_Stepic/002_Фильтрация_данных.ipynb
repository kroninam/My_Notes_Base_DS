# Задачи

```python
# У какой доли студентов из датасэта в колонке lunch указано free/reduced?

len(df[df['lunch'] == 'free/reduced']) / len(df)
```

```python
# Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?

s1 = df.loc[df['lunch'] == 'free/reduced', ['math score', 'reading score', 'writing score']].mean()
print('Среднее:\n',df.loc[df['lunch'] == 'free/reduced', ['math score', 'reading score', 'writing score']].mean())

print()

s2 = df.loc[df['lunch'] == 'free/reduced', ['math score', 'reading score', 'writing score']].var()
print('Дисперсия:\n', df.loc[df['lunch'] == 'free/reduced', ['math score', 'reading score', 'writing score']].var())
```

```python
s3 = df.loc[df['lunch'] == 'standard', ['math score', 'reading score', 'writing score']].mean()
print('Среднее:\n',df.loc[df['lunch'] == 'standard', ['math score', 'reading score', 'writing score']].mean())

print()

s4 = df.loc[df['lunch'] == 'standard', ['math score', 'reading score', 'writing score']].var()
print('Дисперсия:\n', df.loc[df['lunch'] == 'standard', ['math score', 'reading score', 'writing score']].var())
```

```python
df_mean_var = pd.DataFrame({'Среднее_бп': s1, 'Среднее_std': s3, 'Дисперсия_бп': s2, 'Дисперсия_std': s4 })
```

# метод query:

```python
# более компактный способ фильтрации, не требует дублирвоания данных

df.query('writing_score > 74')
```

```python
# сравнение со строкой:

df.query('gender == "female"')
```

```python
# комбинирование запросов фильтрации:


df.query('gender == "female" & writing_score > 74')
# df.query('gender == "female" | writing_score > 74')
```

```python
# фильтрация при помощи переменной:

# задаем переменную:
score = 98

# пытаемся отфильтровать по ней (не получается):
# df.query('writing_score > score')

# !!! переменную надо экранировать при помощи @:

df.query('writing_score > @score')
```
