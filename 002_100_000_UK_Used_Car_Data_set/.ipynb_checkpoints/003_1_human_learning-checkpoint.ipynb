## BaseLine

```python
# Перед тем как делать алгоритм нужно посчитать так назвыаемый BaseLine

# это исходная точка

# это то качество модели, та ошибка модели или алгоритма от которго мы будем отталкиваться.
# если мы не можем сделать лучше чем Baseline - то смысла в нашей работе не будет.


# Можно всем машинам, когда они поступают задавать одну цену: train['price'].mean()

# но на сколько хорош будет этот алгоритм??
```

```python
# Наш Baseline - это предположение, что цены для всех машин можно задавать как среднее значение цены по выборке 
```

```python
train['price'].mean()
```

```python
# делаем новую колонку в train и задаем ей значения среднего:

train['price_pred_mean'] = train['price'].mean()
```

```python
train
```

```python
# рассчитываем ошибку err как разницу между средней ценой и ценой каждого кейса
# получаем значение ошибки с плюсом или минусом - задаем новый столбец err

train['err'] = train['price_pred_mean'] - train['price']
```

```python
train
```

```python
# все ошибки берем по модулю, делаем новый столбец с абсолютными:

# это absolute error

train['err_abs'] = (abs(train['err']))
```

```python
train
```

```python
# это метрика - Mean absolute error (MAE). 

# MAE измеряет среднее абсолютное отклонение между прогнозируемыми и фактическими значениями. 
# Она вычисляется как сумма абсолютных разностей между прогнозами и фактическими значениями, 
# деленная на общее количество прогнозов. MAE показывает, насколько близко прогнозируемые 
# значения к фактическим значениям, без учета направления отклонений. 
# Чем меньше значение MAE, тем лучше качество модели.

train['err_abs'].mean()
```

```python
# посчитаем ошибку в процентах - на сколько в процентах мы ошибились при определении цены, относительно действительной цены:

# используя среднее значение:

# получаем относительную ошибку = absolute percentage error 

train['err_abs_pct'] = train['err_abs'] / train['price']
```

```python
train
```

```python
# теперь считаем среднее по этой ошибке:

# это метрика - mean absolute percentage error (MAPE) 

# MAPE измеряет среднее абсолютное отклонение между прогнозируемыми 
# и фактическими значениями, выраженное в процентах от фактических значений. 
# Эта метрика также позволяет оценить точность модели, но в отличие от 
# MAE она учитывает относительные отклонения между прогнозами и 
# фактическими значениями. Метрика MAPE может быть полезна в случаях, 
# когда важно оценить точность прогнозов с учетом масштаба данных. 
# Также стоит отметить, что MAPE может быть неопределенной (или бесконечной), 
# если фактические значения равны нулю. 
# Чем меньше значение MAPE, тем лучше качество модели.

# это наш Baseline - то есть максимальная ошибка которая может быть и ниже нее нет смысла считать что-то

train['err_abs_pct'].mean()
```

```python
# будет ли прогноз лучше если вместо среднего взять медиану??

# об этом скажут метрики MAE и MAPE. Чем они меньше - тем лучше модель.

train['price'].median()
```

```python
# делаем новую колонку в train и задаем ей значения медианы:

train['price_pred_median'] = train['price'].median()
```

```python
train
```

```python
# получаем ошибку относительно медианной цены и реальной цены:

train['err_median'] = train['price_pred_median'] - train['price']
```

```python
# получаем модуль этой ошибки

np.mean(abs(train['err_median']))
```

```python
# создаем столбец для ошибки

train['err_abs_median'] = (abs(train['err_median']))
```

```python
# считаем среднее ошибки (MAE)

train['err_abs_median'].mean()
```

```python
# посчитаем ошибку в процентах - на сколько в процентах мы ошибились при определении цены, относительно действительной цены:

# используя медианное значение:

train['err_abs_pct_median'] = train['err_abs_median'] / train['price']
```

```python
# расчитываем # MAPE

np.mean(train['err_abs_pct_median'])
```

```python
# Всё то же самое можно проделать с помощью специальных функций sklearn:

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# для среднего:
print(mean_absolute_error(train['price'], train['price_pred_mean']))
print(mean_absolute_percentage_error(train['price'], train['price_pred_mean']))

print('*'*20)

# для медианы:
print(mean_absolute_error(train['price'], train['price_pred_median']))
print(mean_absolute_percentage_error(train['price'], train['price_pred_median']))


# сравнивая абсолютную ошибку (в фунтах цены от каждой продажи) и ошибку в процентах - можно выбрать лучшую метрику
# в данном случае - для данной модели - медиана дает меньшее значение ошибки на 6%. Значит используем медиану.
```

```python
# теперь проверим работу алгоритма (модели) на тестовой выборке: (рассчитаем MAE и MAPE и оценим их)

test
```

```python
# добавляем колонку в test - в которой указваем медиану из набора train !
# то есть на train мы определили среднюю цену и задаем ее в test:

test['price_pred_median'] = train['price'].median()
```

```python
test
```

```python
# рассчитанные значения абсолютной ошибки цены в фунтах (MAE) и ошибки в процетах (MAPE): по медиане

print(mean_absolute_error(test['price'], test['price_pred_median']))
print(mean_absolute_percentage_error(test['price'], test['price_pred_median']))


# здесь характерно то, что ошибку мы считали от медианных значений, полученных на выборке train, 
# добавили это значение из train в выборку test
# и уже считаем ошибки (абсолютную (MAE) и в процентах (MAPE)) от цен, которые находятся в выборке test
```

# Алгоритм (или Модель) на основе годовых данных

## На среднем значении

```python
# Идея алгоритма (модели) - предсказываем цену машины как среднее на основе года выпуска
train['price'].corr(train['year'])
```

```python
df = pd.read_csv('unzipped_archive/bmw.csv')
df
```

```python
from sklearn.model_selection import train_test_split

# по умолчанию train_size = 0.75, (test = 0.25)

# по умолчанию выборки train и test получаются всегда разные при их определении, выбираются случайно

# но выборку можно захолдить, передав парамтетр random_state = 42

train, test = train_test_split(df, train_size = 0.75)
```

```python
# группируем данные по годам и считаем среднюю цену для каждого года

train.groupby(['year'])['price'].agg(['mean', 'median', 'count'])
```

```python
df1 = train.groupby(['year'])['price'].agg(['mean', 'median', 'count'])
```

```python
df1.plot(marker = '+')
```

```python
# интересно, что train.groupby(['year'])['price'].mean() - дает нам Series
# а если добавить .reset_index() - получаем DataFrame


train.groupby(['year'])['price'].mean().reset_index()
```

```python
# задаем нашей модели (или алгоритму) переменную model_year_mean
# это наша модель - то что умеет определять целевую переменную (здесь цена) на основе каких-то параметров (здесь средняя цена)

# это модель для предсказания цены на основе срденей годовой цены:

model_year_mean = train.groupby(['year'])['price'].mean().reset_index()
```

```python
# теперь столбец price - это уже не просто цена, а предсказательный атрибут. 

# Зададим ему имя price_pred_year_mean

model_year_mean.columns = ['year', 'price_pred_year_mean']
```

```python
model_year_mean
```

```python
# Если мы берем корреляцию по всему train - год от цены - то получаем меньшую корреляцию
train['year'].corr(train['price'])
```

```python
# а если мы берем корреляцию год от средней цены за каждый год - получаем корреляцию больше на 20%:
model_year_mean['year'].corr(model_year_mean['price_pred_year_mean'])
```

```python
# теперь можно соединить нашу модель 'model_year_mean' c выборкой 'train':

# how = 'left' - потому что мы хотим чтобы все значения в train остались в итоговом df

# средние для года 'model_year_mean' добавляются для каждой записи в train

pd.merge(train, model_year_mean, how = 'left', on = 'year')
```

```python
# сохраняем train c новым столбцом 'model_year_mean' в переменную train (обновляем train)

train = pd.merge(train, model_year_mean, how = 'left', on = 'year')
```

```python
# теперь надо посчитать ошибки (MAE и MAPE)
```

```python
train['err'] = abs(train['price'] - train['price_pred_year_mean'])
```

```python
# MAE:

train['err'].mean()
```

```python
train['err_percent'] = train['err']/train['price']
```

```python
# MAPE:

train['err_percent'].mean()
```

```python
# Проверка с автоматическим расчетом в sklearn:

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# для среднего:
print(mean_absolute_error(train['price'], train['price_pred_year_mean']))
print(mean_absolute_percentage_error(train['price'], train['price_pred_year_mean']))
```

```python
# теперь ошибки MAE и MAPE надо расчитать на тестовой выборке 'test'

# для этого мы уже к 'test' добавляем нашу модель 'model_year_mean': (модель 'model_year_mean' обучена на обучающей выборке)

test = pd.merge(test, model_year_mean, how = 'left', on = 'year')
```

```python
test
```

```python
# рассчитываем MAE и MAPE вручную:

test['err'] = abs(test['price'] - test['price_pred_year_mean'])
```

```python
# MAE:

test['err'].mean()
```

```python
test['err_percent'] = test['err'] / test['price']
```

```python
# MAPE:

test['err_percent'].mean()
```

```python
# Проверка с автоматическим расчетом в sklearn:

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# для среднего:
print(mean_absolute_error(test['price'], test['price_pred_year_mean']))
print(mean_absolute_percentage_error(test['price'], test['price_pred_year_mean']))
```

```python
# слабые различия MAE и MAPE на train и на test выборках показывают, что алгоритм работает стабильно
```

## Теперь на медиане

```python
df = pd.read_csv('unzipped_archive/bmw.csv')
df
```

```python
from sklearn.model_selection import train_test_split

# по умолчанию train_size = 0.75, (test = 0.25)

# по умолчанию выборки train и test получаются всегда разные при их определении, выбираются случайно

# но выборку можно захолдить, передав парамтетр random_state = 42

train, test = train_test_split(df, train_size = 0.75)
```

```python
# группируем данные по годам и считаем медианную цену для каждого года

train.groupby(['year'])['price'].agg(['mean', 'median', 'count'])
```

```python
# группируем данные по годам и считаем медианную цену для каждого года
train.groupby(['year'])['price'].median().reset_index()
```

```python
# даем название нашей модели (алгоритму) и присваеваем новое значение колонке с ценой 

model_pred_median = train.groupby(['year'])['price'].median().reset_index()
model_pred_median.columns = ['year', 'pred_median']
model_pred_median
```

```python
model_pred_median['year'].corr(model_pred_median['pred_median'])
```

```python
# мержем нашу модель к train

train = pd.merge(train, model_pred_median, how = 'left', on = 'year')
```

```python
train
```

```python
# расчет ошибки (для каждого значения в train)

train['err'] = abs(train['price'] - train['pred_median'])
```

```python
# расчет метрики MAE:

train['err'].mean()
```

```python
# расчет процентной ошибки (для каждого значения в train):

train['err_percent'] = train['err'] / train['price']
```

```python
# расчет метрики MAPE:

train['err_percent'].mean()
```

```python
# Проверка с автоматическим расчетом в sklearn:

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# для среднего:
print(mean_absolute_error(train['price'], train['pred_median']))
print(mean_absolute_percentage_error(train['price'], train['pred_median']))
```

```python
# проверим на тестовой выборке:
```

```python
test
```

```python
# мержим нашу модель к тестовой выборке:

# Проверка с автоматическим расчетом в sklearn не работает естли в test есть nan - без dropna() - не работает.
# однако в ручном режиме nan игнорируется.
# результаты в обоих случаях сходятся.

test = pd.merge(test, model_pred_median, how = 'left', on = 'year').dropna()
```

```python
test
```

```python
test[test['pred_median'].isna()]
```

```python
# считаем ошибку для каждого значения в тестовой выборке:

test['err'] = abs(test['price'] - test['pred_median'])
```

```python
# считаем MAE:

test['err'].mean()
```

```python
# считаем ошибку в процентах для каждого значения в тестовой выборке:

test['err_percent'] = test['err'] / test['price']
```

```python
# считаем MAPE:

test['err_percent'].mean()
```

```python
# Проверка с автоматическим расчетом в sklearn: (работает если дропнуть nan - несколько шагов назад)

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# для медианы:
print(mean_absolute_error(test['price'], test['pred_median']))
print(mean_absolute_percentage_error(test['price'], test['pred_median']))
```
