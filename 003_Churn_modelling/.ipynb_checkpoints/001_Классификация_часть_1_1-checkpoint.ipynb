# Метрика Uplift

```python
# Метрика uplift поднимает большие значения прогноза вверх, после чего их можно снимать как сливки 
# и по этим сливкам смотеть самые интересные данные о целевом признаке. 

# И проверять как наш алгоритм поднимает эти сливки наверх - хорошо или не очень.

# в данном случае о количестве ушедших.

# сначала снимается например 20% случайной выборки (sample), а затем производится сортировка по нашему прогнозному признаку
# (в данном случае это score_prod) и берется 20% сливок сверху. Потом делится первое на второе

# В результате получаем отношение данных о целевом признаке по "сливкам" (uplift - данным) к рандомному выбору.

# Таким образом получаем во сколько раз наш прогноз лучше рандомного прогноза.
# во сколько раз обгоняется по точности рандомная модель
```

```python
train_val
```

```python
# Возьмем из train_val['Exited'] 30 % (frac = 0.3) случайных людей: и посчитаем среднее:

# вообще сколько бы процентов случайных строк мы не взяли значние mean всегда будет около score_mean по закону больших чисел:
# среднее в выборке стремится к среднему во всем df
    
print(train_val['Exited'].sample(frac = 0.2, random_state = 42).mean())
print(train_val['Exited'].sample(frac = 0.3, random_state = 42).mean())
print(train_val['Exited'].sample(frac = 0.4, random_state = 42).mean())
print(train_val['Exited'].sample(frac = 0.5, random_state = 42).mean())
print(train_val['Exited'].sample(frac = 0.6, random_state = 42).mean())
```

```python
# сколько всего в выборке клиентов:

train_val['Exited'].count()
```

```python
# посчитаем сколько всего в выборке отточных клиентов:

exited_all = train_val['Exited'].sum()
exited_all
```

```python
# посчитаем сколько в случайной выборке отточных клиентов: (например на 30% - frac = 0.3)

exited_30 = train_val['Exited'].sample(frac = 0.3, random_state = 42).sum()
exited_30
```

```python
# а теперь поделим сколько в случайной выборке отточных клиентов на общее кол-во в этой выборке:

# сколько процентов выборки берем (frac = 0.3) - столько процентов отточников и получаем. профита нет

exited_30 / exited_all
```

```python
# А теперь отсортируем всю выборку train_val по scor_prod (score сделанный на основе NumOfProducts):

train_val_sorted = train_val.sort_values(by = 'score_prod', ascending = False)
train_val_sorted
```

```python
# теперь надо взять топ-20% этой отсортированной выборки:

# получаем число - количество записей которые будут соответствовать 20% всей выборки:
sorted_20 = round(len(train_val) * 0.2)

# получаем топ-20% отобранных по score_prod:
train_val_sorted.head(sorted_20)
```

```python
# посчитаем сколько отточников нашлось в этих сортированных по score_prod строках:

train_val_sorted.head(sorted_20)['Exited'].sum()
```

```python
# и процент отточников в этой выборке:

train_val_sorted.head(sorted_20)['Exited'].sum() / exited_all
```

```python
# UPLIFT метрика находится так:

# процент отточников в сортированной по SCORE выборке деленный на процент отточников в произвольной выборке:

# процент отточников в сортированной по SCORE выборке:
sorted_20 = round(len(train_val) * 0.2)
top_20_score_prod = train_val_sorted.head(sorted_20)['Exited'].sum() / exited_all


# процент отточников в НЕ-сортированной выборке:
exited_20 = train_val['Exited'].sample(frac = 0.2, random_state = 42).sum() / exited_all


# метрика uplift: показывает во сколько раз сортировка по SCORE лучше чем случайная сортировка

top_20_score_prod / exited_20
```

# Добавить шум

```python
# чтобы продемонстрировать, что для uplift важен именно порядок ранжирования

#### в принципе в uplift можно и не добавлять шум. 
# Шум добавляется для того, чтобы все значения столбца score_prod (нашего прогноза) - стали уникальными
# и более красиво сортировались (ранжировались) в вверху выборки (без шума - было бы много одинаковых значений вверху)

### Кроме красивости добавление шума и сортировка по уникальным значениям SCORE становится воспроизводимой - 
# то есть сортировка будет всегда одинаковой (т.к. уникальные значения SCORE, а не все одинаковые)
```

```python
import numpy as np
```

```python
# добавляет произвольный шум, каждый раз новый:

np.random.uniform(0, 0.001, size = len(train_val))
```

```python
# добавляет произвольный шум, один раз и постоянно один и тот же: (чтобы всегда иметь один и тот же результат)

np.random.seed(42)
np.random.uniform(0, 0.001, size = len(train_val))
```

```python
# добавляем (точенее вычетаем) фиксированный шум из score_prod:

np.random.seed(42)
train_val['score_prod'] = train_val['score_prod'] - np.random.uniform(0, 0.001, size = len(train_val))
```

```python
# теперь каждый из значений score_prod - стал уникальным:

train_val['score_prod'].nunique()
```

# Создаем функцию для UPLIFT

```python
# Чтобы в дальнейшем было удобно считать uplift - создаем функцию: (из всех шагов рассмотренных выше)


# функция принмает df, определенный столбец со SCORE и pct - это сколько процентов выборки мы берем:
def uplift(df, score, pct):
    # посчитаем сколько всего в выборке отточных клиентов:
    exited_all = df['Exited'].sum()
    
    # отсортируем df по SCORE:
    df_sorted = df.sort_values(by = score, ascending = False)
    
    
    # теперь надо взять топ-20% этой отсортированной выборки:

    # получаем количество записей которые будут соответствовать pct % всей выборки:
    sorted_20 = round(len(df) * pct)
    
    # посчитаем сколько отточников нашлось в этих сортированных по score_prod строках:
    exited_found = df_sorted.head(sorted_20)['Exited'].sum()
    
    return (exited_found / exited_all) / pct    
    
```

```python
# проверим функцию uplift на score_prod: (здесь не сходится с тем что выше т.к. добавили шум)

uplift(train_val, 'score_prod', 0.2)
```
